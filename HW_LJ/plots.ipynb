```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_pos_small = pd.read_csv("data/positions_small.csv")

plt.scatter(df_pos_small['x0'][200:], df_pos_small['y0'][200:], c='blue')
plt.scatter(df_pos_small['x1'][200:], df_pos_small['y1'][200:], c='red')

plt.show()
```

```python
df_energy_small = pd.read_csv("data/verification_small.csv")

plt.plot(df_energy_small['step'], df_energy_small['Epot'], label='Epot')
plt.plot(df_energy_small['step'], df_energy_small['Ekin'], label='Ekin')
plt.plot(df_energy_small['step'], df_energy_small['Etot'], label='Etot')
plt.legend()
plt.show()
```
